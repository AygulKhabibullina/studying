<h2 style="text-align: center;"><b>Библиотека seaborn</b></h2> 

**Seaborn** - библиотека визуализации данных в Python основанная на `matplotlib`, т.е. используется тот же `matplotlib`.

Рассмотрим пример анализа данных катострофы Титаника с помощью визуализации в [seaborn](https://seaborn.pydata.org/index.html).   

```python
import numpy as np
import pandas as pd

from matplotlib import pyplot as plt
```

# Титаник

```python
url = "https://raw.github.com/mattdelhey/kaggle-titanic/master/Data/train.csv"
titanic = pd.read_csv(url)
titanic.info()
```

```python
titanic.head()
```

```python
titanic.describe()
```

# Количественные признаки

`data.hist()` строит гистограмму по данным с помощью seaborn, если он был импортирован

```python
titanic['age'].hist()
plt.show()
```

```python
import seaborn as sns
```

```python
titanic['age'].hist()
plt.show()
```

* `sns.set(style)` - настройка стиля: **darkgrid**, **whitegrid**, **dark**, **white**, **ticks**

```python
sns.set(style="darkgrid")
```

```python
titanic['age'].hist()
plt.show()
```

* `sns.kdeplot(data)` - строит сглаженное распределение данных

```python
sns.kdeplot(titanic["age"].dropna())
```

```python
sns.kdeplot(titanic["age"].dropna(), shade=True)
```

* `sns.distplot(data)` - hist() + kdeplot()

```python
sns.histplot(titanic["age"].dropna(),  bins=10, kde=True, stat="density", linewidth=0)
plt.show()
```

* `sns.boxplot(data)` - построение коробочки со статистиками: 
    - **линия по центру** - медиана
    - **границы коробки** - 25 и 75 перцентиль (пространство между ними - интерквартильный размах) 
    - **усы** - разброс данных, кроме выбросов
    - **точки** - выбросы

```python
sns.boxplot(x="age", data=titanic)
```

# Категориальные признаки

```python
titanic['sex'].value_counts()
```

* `sns.countplot(x, **kwargs)` - строит bar по данным
        x - сами данные или имя столбца для data
        data - таблица с данными
        hue - для каких категорий разделить нужный признак
    

```python
sns.countplot(x="sex", data=titanic)
```

```python
sns.countplot(x="age", data=titanic)
```

```python
sns.countplot(x="pclass", data=titanic)
```

# Количественные vs Количественные

```python
titanic.hist()
plt.show()
```

# Категориальные vs Категориальные

```python
sns.countplot(x="pclass", data=titanic, hue="sex")
```

```python
sns.countplot(x="pclass", data=titanic[titanic['age'] > 50], hue="sex")
```
