# Задание 1.1

Изобразите график функции по заданным точкам x и y цветом 'magenta'.

Изобразите сетку синим цветом точечной линией толщиной 0.3.

Напишите "легенду" графика, подпишите оси, дайте название графику. 

Изобразите график в диапазоне x от -10 до 10  и y от -10 до 10, и пусть шкала идет для каждого второго целого значения (-10, -8, -6 ...).

```python
t = np.linspace(0, 60, 10000)

x=4.4*(np.cos(t)+np.cos(1.1*t)/1.1)
y=4.4*(np.sin(t)-np.sin(1.1*t)/1.1) 


a = -10
b = 10

plt.figure(figsize=(7, 7))

plt.grid(ls=':', color='b')
plt.plot(x, y, 'c', label='График',  linewidth=0.3)

plt.legend(fontsize=14, shadow=True, loc='upper left')

plt.xlabel('Ось x', fontsize=15)
plt.ylabel('Ось y', fontsize=15)
plt.xticks(np.linspace(a, b, 9), 
           fontsize=10)
plt.yticks(fontsize=12)
plt.xlim(a, b)
plt.ylim(a, b)

plt.title('Графики matplotlib', fontsize=15)

plt.show()
```

# Задание 2

Изобразите 4 графика (2 х 2) для функций sin, cos, tan, cot с помощью subplot. (Будьте внимательны, в некоторых точках функции могут не существовать).

Красиво оформите эти графики, сделайте все подписи, подберите размеры так, чтобы надписи разных графиков не перекрывались.

```python
x = np.linspace(-10, 10, 1000)
```

```python
grid = plt.GridSpec(2, 2, wspace=0.4, hspace=0.2)

def cot(x):
  cot = np.cos(x)/np.sin(x)
  cot[:-1][np.diff(cot) > 0] = np.nan
  return cot

def tan(x):
  tan = np.tan(x)
  tan[:-1][np.diff(tan) < 0] = np.nan
  return tan

plt.figure(figsize=(12, 12))

plt.subplot(grid[0, 0])
plt.gca().set_title('sin', fontsize=24)
plt.plot(x, np.sin(x))
plt.grid(ls=':', color='b', linewidth=0.3)

plt.subplot(grid[0, 1], label='cos')
plt.gca().set_title('cos', fontsize=24)
plt.plot(x, np.cos(x))
plt.grid(ls=':', color='b', linewidth=0.3)

plt.subplot(grid[1, 0])
plt.gca().set_title('tan', fontsize=24)
plt.plot(x, tan(x))
plt.ylim(-25, 25)
plt.grid(ls=':', color='b', linewidth=0.3)

plt.subplot(grid[1, 1])
plt.gca().set_title('cot', fontsize=24)
plt.plot(x, cot(x))
plt.ylim(-25, 25)
plt.grid(ls=':', color='b', linewidth=0.3)
```
