[`Matplotlib`](https://matplotlib.org/) — библиотека для визуализации данных. Пакет поддерживает многие виды графиков и диаграмм:
- Графики (line plot)
- Диаграммы рассеяния (scatter plot)
- Столбчатые диаграммы (bar chart) и гистограммы (histogram)
- Круговые диаграммы (pie chart)
- и др.

Получаемые графики можно сохранять в различных форматах: векторных (`eps`, `pdf`, `svg`) и в растровых (`png`, `jpg`)

`matplotlib` позволяет строить графики `2D` очень многих типов с большими возможностями их настройки. 
Также поддерживает небольшую работу с `3D`-графиками.

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Начало

Основной функцией создания графика является `plt.plot()`:  

* `plt.plot(x, y, **kwargs)` - создает график по набору точек `x` и `y`. Первая точка будет иметь координаты (x[0], y[0]), вторая (x[1], y[1]) и т.д.  `**kwargs` - дополнительные параметры функции

Точки будут соединяться прямыми, т.е. строится ломаная линия.

* `plt.show()` - отрисовывает график

```python
x = [1, 2, 3]
y = [0, 7, 2]
plt.plot(x, y)
plt.show()
```

```python
x = [1, 2, 3, 10, 5]
y = [0, 7, 2, 10, 10]
plt.plot(x, y)
plt.show()
```

Мы можем использовать сетку, чтобы видеть примерное местоположение интересующих нас точек

* `plt.grid(**kwargs)` -- нанести сетку с некими параметрами

        linestyle - вид сетки ('--', ':', ...)
        linewidth - ширина линий
        color - цвет
        ...

```python
x = [1, 2, 3, 10, 5]
y = [0, 7, 2, 10, 10]
plt.plot(x, y)
plt.grid()
plt.show()
```

```python
x = [1, 2, 3, 10, 5]
y = [0, 7, 2, 10, 10]
plt.plot(x, y)
plt.grid(linestyle = '--')
plt.show()
```

```python
x = [1, 2, 3, 10, 5]
y = [0, 7, 2, 10, 10]
plt.plot(x, y)
plt.grid(color='g', ls='--', linewidth=0.5)
plt.show()
```

* `plt.axes().set_aspect(1)` -- сделать оси x и y в одинаковой пропорции (тут видно, что получился квадратный рисунок)

```python
x = [1, 2, 3, 10, 5]
y = [0, 7, 2, 10, 10]
plt.axes().set_aspect(1)
plt.plot(x, y)

plt.grid(color='g', ls='--', linewidth=0.5)
plt.show()
```

* `plt.figure(figsize=(width, height))` - если написать это перед `plt.plot()`, то изменится размер итоговой картинки с графиком. Пример:

```python
x = [1, 2, 3, 10, 5]
y = [0, 7, 2, 10, 10]

plt.figure(figsize=(7, 5))
plt.plot(x, y)

plt.grid(color='g', ls='--', linewidth=0.1)
plt.show()
```

## Объектно-ориентированный стиль

До этого мы использовали так называемый `процедурный` стиль построения графиков. Сейчас рассмотрим второй популярный стиль, называемый объектно-ориентированным.

Рисунок `figure` содержит общие настройки графиков. 

Оси `axis` - это объекты, которые содержат наши графики.

* `ax.plot(x, y)` -- построение графика по точкам (x[0], y[0]), ..., (x[n], y[n])

```python
# создание рисунка и осей (графика)
fig = plt.figure()
ax = plt.axes()

x = np.linspace(-10, 10, 1000)
ax.plot(x, np.sin(x))
plt.show()
```

* `ax.axis([x_min, x_max, y_min, y_max])` -- задание границ осей
        

```python
# создание рисунка и осей (графика)
fig = plt.figure()
ax = plt.axes()

ax.axis([-5, 8, -2, 2])
x = np.linspace(-10, 10, 1000)
ax.plot(x, np.sin(x))
```

Мы можем разделять график по частям и в каждой рисовать нужный нам график





```python
fig, ax = plt.subplots(2)

ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
plt.show()
```

Соотвествие между `процедурным` и `объекто-ориентированным` стилями:
- plt.xlabel() → ax.set_xlabel()
- plt.ylabel() → ax.set_ylabel()
- plt.xlim() → ax.set_xlim()
- plt.ylim() → ax.set_ylim()
- plt.title() → ax.set_title()

```python
fig = plt.figure()
ax = plt.axes()

ax.grid(color='grey', linestyle='--', linewidth=0.5)
ax.plot(x, np.sin(x))

ax.set_xlim(-5, 5)
ax.set_ylim(-1.5, 1.5)
ax.set_xlabel('x', fontsize=15)
ax.set_ylabel('sin(x)', fontsize=15)
ax.set_title('Plot', fontsize=25)

plt.show()
```

### Распределения

```python
import scipy.stats as sps
```

```python
#sps.norm.rvs - нормальные непрерывные случаные значения
sample = sps.norm.rvs(size=300) # нормальное

plt.figure(figsize=(10, 1))
plt.grid(ls='--')

plt.scatter(sample, np.zeros(len(sample)), alpha=0.2)#, color = 'r')
plt.show()
```

```python
norm = np.linspace(-3, 3, len(sample))  # точки для нормального распределения

plt.figure(figsize=(10, 4))
plt.grid(ls='--')
#sts.norm.pdf  - функция плотности вероятности (Probability density function) 
plt.scatter(sample, np.zeros(len(sample)) , alpha=0.2, label='данные')
plt.plot(norm, sps.norm.pdf(norm), color='r', label='распределение')
plt.legend()  
plt.show()
```

```python
#sps.expon.rvs - Экспоненциальные непрерывные случайные значения
sample = sps.expon.rvs(size=300) # экспоненциальное

plt.figure(figsize=(10, 1))
plt.grid(ls='--')
plt.scatter(sample, np.zeros(len(sample)), alpha=0.2, color='b')
plt.show()
```

```python
expon = np.linspace(-1, 3, len(sample))  # точки для экспоненциального распределения

plt.figure(figsize=(10, 4))
plt.grid(ls='--')

plt.scatter(sample, np.zeros(len(sample)) , alpha=0.2, label='данные')
plt.plot(expon, sps.expon.pdf(expon), color='r', label='распределение')
plt.legend()  
plt.show()
```

### Статистики на графике

```python
# import numpy as np
# import pylab as plt

N = 3000
total_bins = 5

# Sample data
#np.random.random - случайные вещественные числа
X = np.random.random(size=N)*10
Y = X**2 + np.random.random(size=N)*X*10

bins = np.linspace(X.min(), X.max(), total_bins)

delta = bins[1] - bins[0]

#np.digitize - возвращает индексы ячеек, которым принадлежит каждое значение в массиве
idx  = np.digitize(X, bins) # в какие bins попадают X
#медиана в каждом bins
running_median = [np.median(Y[idx==k]) for k in range(1, total_bins)]
#Вычислить стандартное отклонение по оси
running_std    = [Y[idx==k].std() for k in range(1, total_bins)]

#plt.errorbar - используется для построения графика y в зависимости от x в виде линий и / или маркеров с прикрепленными полосами ошибок
plt.figure(figsize=(15, 5))
plt.scatter(X, Y, color='g', alpha=.2, s=2)
plt.plot(bins[1:]-delta/2, running_median, 'r--', lw=4, alpha=.8)
plt.errorbar(bins[1:]-delta/2, running_median, running_std)
plt.axis('tight')
plt.show()
```

## Трёхмерная линия

Задаётся параметрически: $x=x(t)$, $y=y(t)$, $z=z(t)$.

```python
#Создание 3D графиков
from mpl_toolkits.mplot3d import Axes3D
```

```python
t = np.linspace(0, 4 * np.pi, 100)
x = np.cos(t)
y = np.sin(t)
z = t / (4 * np.pi)
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.plot(x, y, z)
plt.show()
```

```python
X = 10
N = 50
u = np.linspace(-X, X, N)
# np.meshgrid -Возвращает координаты матрицы из векторов
x, y = np.meshgrid(u, u)
z = x**2 + y**2 - 2*y

fig = plt.figure()
ax = Axes3D(fig)
# rstride - Шаг строки массива (размер шага)
#cstride - Шаг столбца массива (размер шага)
ax.plot_surface(x, y, z, rstride=1, cstride=1)
plt.show()
```

```python

X = 10
N = 50
u = np.linspace(-X, X, N)
x, y = np.meshgrid(u, u)
r = np.sqrt(x ** 2 + y ** 2)
z = np.sin(r) / r

fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(x, y, z, rstride=1, cstride=1)
plt.show()
```
